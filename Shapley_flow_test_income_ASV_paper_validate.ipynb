```python
import os
os.environ["OMP_NUM_THREADS"] = "8"
os.environ["OPENBLAS_NUM_THREADS"] = "8"
os.environ["MKL_NUM_THREADS"] = "8"
os.environ["VECLIB_MAXIMUM_THREADS"] = "8"
os.environ["NUMEXPR_NUM_THREADS"] = "8"

# standard libaries
import sys
import copy
import dill
from collections import defaultdict
import numpy as np
import pandas as pd
import xgboost
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import shap

# flow specific packages
if '../' not in sys.path:
    sys.path = ['../'] + sys.path
from flow import Node, CreditFlow, Graph, get_source_nodes, topo_sort, flatten_graph, eval_graph, boundary_graph, single_source_graph, viz_graph, save_graph, hcluster_graph
from flow import ParallelCreditFlow, GraphExplainer, translator
from flow import group_nodes, build_feature_graph
from flow import CausalLinks, create_xgboost_f, create_linear_f
from flow import edge_credits2edge_credit, node_dict2str_dict
from on_manifold import OnManifoldExplainer, IndExplainer

np.random.seed(42) # fix for reproducibility
```

## Dataset

```python
X,y = shap.datasets.adult()
X_display,y_display = shap.datasets.adult(display=True) # human readable feature values

# create a train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
xgb_train = xgboost.DMatrix(X_train, label=y_train)
xgb_test = xgboost.DMatrix(X_test, label=y_test)

print(X.shape)
X_display.head()
```

## Training

```python
params = {
    "eta": 0.01,
    "max_depth": 2,
    'objective':'binary:logistic', # output probability
    "subsample": 0.9
}
model = xgboost.train(params, xgb_train, 3000, evals = [(xgb_test, "test")], verbose_eval=1000)

# workaround from https://github.com/slundberg/shap/issues/1215
mybooster = model

model_bytearray = mybooster.save_raw()[4:]
def myfun(self=None):
    return model_bytearray
mybooster.save_raw = myfun
```

## (1) SHAP value
SHAP value assumes all features are indpendent

```python
explainer = shap.TreeExplainer(mybooster)
shap_values = explainer.shap_values(X_train)
shap_values = pd.DataFrame(shap_values)
shap_values.columns = X_train.columns
df_shap = pd.DataFrame(shap_values.sum(axis = 0, skipna = True)).sort_values(by=[0],ascending=False,key=abs)
```

```python
## plot
shap.summary_plot(shap_values, X_train, plot_type="bar")
```

## Define causal graph for ASV & shapley flow
Causal graph is similar to ASV paper, {age, sex, country, race} are causes to remaining features

```python
## setting up enviroment for graph viz
os.environ["PATH"] += os.pathsep + 'C:/Users/vivianjiang/anaconda3/pkgs/graphviz-2.38-hfd603c8_2/Library/bin'
```

```python
## Complex causal graph
causal_links = CausalLinks()
categorical_feature_names = ['Workclass', 'Marital Status', 'Occupation',
                             'Relationship', 'Race', 'Sex', 'Country']
display_translator = translator(X.columns, X, X_display)
target_name = 'predicted income >50k' # 'predicted hazard'
feature_names1 = ['Marital Status', 'Workclass','Education-Num', 'Occupation','Relationship','Capital Gain','Capital Loss', 'Hours per week']
f = lambda x: x.dot(model.coef_.reshape(-1, 1)).ravel() + model.intercept_ # logit is linear log(p / 1-p)

# build causal links
causal_links.add_causes_effects(feature_names1, target_name, create_linear_f(feature_names1, f))
causal_links.add_causes_effects('Age', 'Marital Status', lambda x: x)
causal_links.add_causes_effects('Sex', 'Marital Status', lambda x: x)
causal_links.add_causes_effects('Country', 'Marital Status', lambda x: x)
causal_links.add_causes_effects('Race', 'Marital Status', lambda x: x)

causal_links.add_causes_effects('Age', 'Workclass', lambda x: x)
causal_links.add_causes_effects('Sex', 'Workclass', lambda x: x)
causal_links.add_causes_effects('Country', 'Workclass', lambda x: x)
causal_links.add_causes_effects('Race', 'Workclass', lambda x: x)

causal_links.add_causes_effects('Age', 'Education-Num', lambda x: x)
causal_links.add_causes_effects('Sex', 'Education-Num', lambda x: x)
causal_links.add_causes_effects('Country', 'Education-Num', lambda x: x)
causal_links.add_causes_effects('Race', 'Education-Num', lambda x: x)

causal_links.add_causes_effects('Age', 'Occupation', lambda x: x)
causal_links.add_causes_effects('Sex', 'Occupation', lambda x: x)
causal_links.add_causes_effects('Country', 'Occupation', lambda x: x)
causal_links.add_causes_effects('Race', 'Occupation', lambda x: x)

causal_links.add_causes_effects('Age', 'Relationship', lambda x: x)
causal_links.add_causes_effects('Sex', 'Relationship', lambda x: x)
causal_links.add_causes_effects('Country', 'Relationship', lambda x: x)
causal_links.add_causes_effects('Race', 'Relationship', lambda x: x)

causal_links.add_causes_effects('Age', 'Capital Gain', lambda x: x)
causal_links.add_causes_effects('Sex', 'Capital Gain', lambda x: x)
causal_links.add_causes_effects('Country', 'Capital Gain', lambda x: x)
causal_links.add_causes_effects('Race', 'Capital Gain', lambda x: x)

causal_links.add_causes_effects('Age', 'Capital Loss', lambda x: x)
causal_links.add_causes_effects('Sex', 'Capital Loss', lambda x: x)
causal_links.add_causes_effects('Country', 'Capital Loss', lambda x: x)
causal_links.add_causes_effects('Race', 'Capital Loss', lambda x: x)

causal_links.add_causes_effects('Age', 'Hours per week', lambda x: x)
causal_links.add_causes_effects('Sex', 'Hours per week', lambda x: x)
causal_links.add_causes_effects('Country', 'Hours per week', lambda x: x)
causal_links.add_causes_effects('Race', 'Hours per week', lambda x: x)


causal_graph = build_feature_graph(X.fillna(X.mean()), causal_links, categorical_feature_names
                                   , display_translator, target_name, method='linear')

causal_graph.draw()
```

## Explanation settings

```python
n_bg = 10 # number of sampled background samples
nsamples = 1000 # number of fg samples to explain
nruns = 10 # paper: 10,000, number of monte carlo samplings
bg = X.fillna(X.mean()).sample(n_bg) # background samples
fg = X[:nsamples] # foreground samples to explain
sample_ind = 5 # sample to show
```

```python
# multiple background result
edge_credits = [] # save for each run
for i in range(len(bg)): # average run result using linearity axiom
    print(f'bg {i}/{len(bg)}')
    explainer = GraphExplainer(causal_graph, bg[i:i+1], nruns=nruns) 
    cf = explainer.shap_values(fg)
    edge_credits.append(node_dict2str_dict(cf.edge_credit)) # use node name is more portable vs. node
```

```python
cf.draw(sample_ind, max_display=10, edge_credit=edge_credits2edge_credit(edge_credits, cf.graph), show_fg_val=False)
```

## On manifold SHAP

```python
feature_names = X.columns
f = lambda x: model.predict(xgboost.DMatrix(pd.DataFrame(x, columns=feature_names))) # wrapper for the model used by on manifold explainer

# multiple background result
manifold_edge_credits = [] # save for each run
bg_on_manifold = X.fillna(X.mean()).sample(100) # used to learn conditional dependence
for i in range(len(bg)): # average run result using linearity axiom 
    print(f'bg {i}/{len(bg)}')
    explainer = OnManifoldExplainer(f, pd.concat([bg[i:i+1], bg_on_manifold]), nruns=nruns)
    cf_on_manifold = explainer.shap_values(fg)
    manifold_edge_credits.append(cf_on_manifold) # use node name is more portable vs. node
```

```python
print('On manifold')
sum(manifold_edge_credits).draw(sample_ind, show=False, max_display=10)
plt.tight_layout()
plt.show()
```

## Shapley Flow

```python
causal_links = CausalLinks()
categorical_feature_names = ['Workclass', 'Marital Status', 'Occupation',
                             'Relationship', 'Race', 'Sex', 'Country']
display_translator = translator(X.columns, X, X_display)
target_name = 'predicted income >50k'
feature_names = list(X.columns)

# build causal links
causal_links.add_causes_effects(feature_names, target_name, create_xgboost_f(feature_names, model)) # target is probability
A = ['Age', 'Sex', 'Country', 'Race']
D = list(set(feature_names) - set(A))
causal_links.add_causes_effects(A, D)

causal_graph = build_feature_graph(X.fillna(X.mean()), causal_links, categorical_feature_names, display_translator, target_name, method='linear')
causal_graph.draw()
```

```python
# multiple background result
causal_edge_credits = [] # save for each run
for i in range(len(bg)): # average run result using linearity axiom
    print(f'bg {i}/{len(bg)}')
    cf_c = GraphExplainer(causal_graph, bg[i:i+1], nruns=nruns).shap_values(fg)
    causal_edge_credits.append(node_dict2str_dict(cf_c.edge_credit)) # use node name is more portable vs. node
```

```python
cf_c.draw(sample_ind, max_display=10, show_fg_val=False, edge_credit=edge_credits2edge_credit(causal_edge_credits, cf_c.graph))
```
